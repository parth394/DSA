```python
# Topological Sorting

def topoSort(graph):
    V = len(graph)
    visit = [False] * V
    stack = []

    for i in range(V):
        if visit[i]==False:
            dfsRec(i, graph, visit, stack)

    for ele in stack[::-1]:
        print(ele)

def dfsRec(u, graph, visit, stack):
    visit[u]=True

    for v in graph:
        if visit[v] ==False:
            dfsRec(v, graph, visit, stack)
    stack.append(u)
```
